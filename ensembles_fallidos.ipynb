```r
require( "data.table" )

setwd("/content/buckets/b1/exp")
# Carga de predicciones de probabilidad (202108)
f1 <- fread("./M1/prediccion.txt")[, .(numero_de_cliente, foto_mes, prob1 = prob)]
f2 <- fread("./M2/prediccion.txt")[, .(numero_de_cliente, foto_mes, prob2 = prob)]
f3 <- fread("./M3/prediccion.txt")[, .(numero_de_cliente, foto_mes, prob3 = prob)]
f4 <- fread("./M4/prediccion.txt")[, .(numero_de_cliente, foto_mes, prob4 = prob)]
f5 <- fread("./M5/prediccion.txt")[, .(numero_de_cliente, foto_mes, prob5 = prob)]
f6 <- fread("./M6/prediccion.txt")[, .(numero_de_cliente, foto_mes, prob6 = prob)]
f7 <- fread("./M7/prediccion.txt")[, .(numero_de_cliente, foto_mes, prob7 = prob)]
f8 <- fread("./M8/prediccion.txt")[, .(numero_de_cliente, foto_mes, prob8 = prob)]
f9 <- fread("./M9/prediccion.txt")[, .(numero_de_cliente, foto_mes, prob9 = prob)]
f10 <- fread("./M10/prediccion.txt")[, .(numero_de_cliente, foto_mes, prob10 = prob)]
f11 <- fread("./apo-506-eze1/prediccion.txt")[, .(numero_de_cliente, foto_mes, prob11 = prob)]
f12 <- fread("./apo506-agus/prediccion-3.txt")[, .(numero_de_cliente, foto_mes, prob12 = prob)]
f13 <- fread("./apo-506-prediccion_eze2/prediccion.txt")[, .(numero_de_cliente, foto_mes, prob13 = prob)]


# # 1. Ganancia máxima para cada modelo en 202107
# g_m1 <- 421600000
# g_m2 <- 421600000
# g_m3 <- 433440000
# g_m4 <- 433440000
# g_m5 <- 436800000
# g_m6 <- 436800000
# g_m7 <- 435200000
# g_m8 <- 430800000
# g_m9 <- 433200000
# g_m10 <- 428000000

g_m1 <- 0
g_m2 <- 0
g_m3 <- 0
g_m4 <- 0
g_m5 <- 0
g_m6 <- 0
g_m7 <- 0
g_m8 <- 0
g_m9 <- 0
g_m10 <- 0
g_m11 <- 0
g_m12 <- 1
g_m13 <- 1

# Definir los modelos A ENSAMBLAR y sus ganancias
ganancias_a_ensamblar <- c(g_m1, g_m2, g_m3, g_m4, g_m5, g_m6, g_m7, g_m8, g_m9, g_m10, g_m11, g_m12, g_m13)

# Calcular la suma total de las ganancias de los modelos seleccionados
suma_total_ganancia <- sum(ganancias_a_ensamblar)

# 4. Calcular los pesos de ponderación
w1 <- g_m1/suma_total_ganancia
w2 <- g_m2/suma_total_ganancia
w3 <- g_m3/suma_total_ganancia
w4 <- g_m4/suma_total_ganancia
w5 <- g_m5/suma_total_ganancia
w6 <- g_m6/suma_total_ganancia
w7 <- g_m7/suma_total_ganancia
w8 <- g_m8/suma_total_ganancia
w9 <- g_m9/suma_total_ganancia
w10 <- g_m10/suma_total_ganancia
w11 <- g_m11/suma_total_ganancia
w12 <- g_m12/suma_total_ganancia
w13 <- g_m13/suma_total_ganancia

# cat("--- Pesos de Ponderación (Basados en Ganancia Máxima Test 202106) ---\n")
# cat(sprintf("Peso M1 seed 100003 (w1): %.4f\n", w1))
# cat(sprintf("Peso M2 seed 990013 (w2): %.4f\n", w2))
# cat(sprintf("Peso M3 seed 100003 (w3): %.4f\n", w3))
# cat(sprintf("Peso M4 seed 990013 (w4): %.4f\n", w4))
# cat(sprintf("Peso M5 seed 100003 (w5): %.4f\n", w5))
# cat(sprintf("Peso M6 seed 990013 (w6): %.4f\n", w6))
# cat(sprintf("Peso M7 seed 100003 (w7): %.4f\n", w7))
# cat(sprintf("Peso M8 seed 100003 (w8): %.4f\n", w8))
# cat(sprintf("Peso M9 seed 100003 (w9): %.4f\n", w9))
# cat(sprintf("Peso M10 seed 990013 (w10): %.4f\n", w10))

cat(sprintf("Suma de Pesos: %.4f\n", w1 + w2 + w3 + w4 + w5 + w6 + w7 + w8 + w9 + w10 + w11 + w12 + w13))
cat("----------------------------------------------------------------\n")


# Lista de Data.Tables a ensamblar
tablas_a_ensamblar <- list(f1, f2, f3, f4, f5, f6, f7, f8, f9, f10, f11, f12, f13)

# Combinar las probabilidades
ensamble <- Reduce(function(x, y) merge(x, y, by=c("numero_de_cliente","foto_mes")),
                   tablas_a_ensamblar)

# Calcular la probabilidad ponderada (Weighted Ensemble)
ensamble[, prob_ponderada := (w1 * prob1) + (w2 * prob2) + (w3 * prob3) + (w4 * prob4) + (w5 * prob5) + (w6 * prob6) + (w7 * prob7) + (w8 * prob8) + (w9 * prob9) + (w10 * prob10) + (w11 * prob11) + (w12 * prob12) + (w13 * prob13)]

# Preparar el resultado final (ranking)
prediccion_202109_ensemble_ponderado <- ensamble[, .(numero_de_cliente, foto_mes, prob = prob_ponderada)]

cat("Predicción ensemble ponderada (202109) calculada con éxito.\n")
```
